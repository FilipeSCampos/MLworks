```python
transacoes=[]
for i in range(0,450):
    transacoes.append([str(crime.values[i,j]) for j in range(0,16)])
```

```python
regras=apriori(transacoes, min_support=0.003,min_confidence=0.9,min_lift=3, min_lenght=2)
```

```python
#Visuallizando os resultados
resultado=list(regras)
resultado
```

```python
resultado2=[list(x) for x in resultado]
resultado2
```

```python
#Quantitativo de associações identificadas
len(resultado2)
```

```python
#Visualizando as 7 principais regras de associação gerada
reg_associadas=[]
for j in range(0,10):
    reg_associadas.append([list(x) for x in resultado2[j][2]])
```
